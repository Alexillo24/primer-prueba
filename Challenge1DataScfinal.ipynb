### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
print(tienda.dtypes)
print(tienda2.dtypes)
print(tienda3.dtypes)
print(tienda4.dtypes)
```

```python
tienda['Tienda'] = 'tienda_1'
tienda2['Tienda'] = 'tienda_2'
tienda3['Tienda'] = 'tienda_3'
tienda4['Tienda'] = 'tienda_4'

tiendas_unidas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

display(tiendas_unidas.head())
display(tiendas_unidas.tail())
tiendas_unidas.head()
```

#1. Análisis de facturación



```python
ventas_por_tienda = tiendas_unidas.groupby(['Tienda'])['Precio'].sum().reset_index()
display(ventas_por_tienda)
```

```python
# @title Tienda vs Precio

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(ventas_por_tienda['Tienda'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(ventas_por_tienda, x='Precio', y='Tienda', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
costos_por_tienda = tiendas_unidas.groupby(['Tienda'])['Costo de envío'].sum().reset_index()
display(costos_por_tienda)
```

```python
# @title Tienda vs Costo de envío

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(costos_por_tienda['Tienda'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(costos_por_tienda, x='Costo de envío', y='Tienda', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

# 2. Ventas por categoría

```python
Venta_por_tienda_Categoria = tiendas_unidas.groupby(['Tienda', 'Categoría del Producto'])['Precio'].sum().reset_index()
display(Venta_por_tienda_Categoria)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_por_tienda = tiendas_unidas.groupby(['Tienda'])['Calificación'].mean().reset_index()

display(calificacion_promedio_por_tienda)
```

```python
# @title Tienda vs Calificación

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(calificacion_promedio_por_tienda['Tienda'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(calificacion_promedio_por_tienda, x='Calificación', y='Tienda', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

# 4. Productos más y menos vendidos

```python
Venta_por_tienda_TOP_producto = tiendas_unidas.groupby(['Tienda', 'Producto'])['Precio'].sum().reset_index()
display(Venta_por_tienda_TOP_producto)
```

```python
top_5_productos = Venta_por_tienda_TOP_producto.groupby('Tienda').apply(lambda x: x.nlargest(5, 'Precio')).reset_index(drop=True)
bottom_5_productos = Venta_por_tienda_TOP_producto.groupby('Tienda').apply(lambda x: x.nsmallest(5, 'Precio')).reset_index(drop=True)

print("Top 5 productos más vendidos por tienda:")
display(top_5_productos)

print("\nTop 5 productos menos vendidos por tienda:")
display(bottom_5_productos)
```

# 5. Envío promedio por tienda

```python
envio_promedio_por_tienda = tiendas_unidas.groupby(['Tienda'])['Costo de envío'].mean().reset_index()

display(envio_promedio_por_tienda)
```

```python
# @title Tienda vs Costo de envío

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(envio_promedio_por_tienda['Tienda'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(envio_promedio_por_tienda, x='Costo de envío', y='Tienda', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
